<h4> Berikut library pada posttest ini </h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

<h4> melihat dataset </h4>

```python
df = pd.read_csv('cars1.csv')
```

```python
df
```

<h4> Pada atribut objek disini harus dirubah dulu ke atribut tipe angka, jadi menggunakan encoding </h4>

```python
varEncoder = OrdinalEncoder()
df['manufacturer_name'] = varEncoder.fit_transform(df[['manufacturer_name']])
df['model_name'] = varEncoder.fit_transform(df[['model_name']])
df['transmission'] = varEncoder.fit_transform(df[['transmission']])
df['color'] = varEncoder.fit_transform(df[['color']])
df['engine_fuel'] = varEncoder.fit_transform(df[['engine_fuel']])
df['engine_has_gas'] = varEncoder.fit_transform(df[['engine_has_gas']])
```

```python
df
```

<h4> disini X sebagai record dan Y sebagai label <h/4>

```python
X = df[['manufacturer_name','model_name','transmission','color','engine_fuel', 'engine_capacity']]
Y = df[['engine_has_gas']]
```

<h4> proses membagi data train </h4>

```python
xTrain, xTest, yTrain, yTest = train_test_split(X,
                                                Y,
                                                test_size=.3)
```

<h4> jumlah kelas ada 2 </h4>

```python
num = 2
yTrain = tf.keras.utils.to_categorical(yTrain, num)
yTest = tf.keras.utils.to_categorical(yTest, num)
```

<h4> Lalu membuat model sequential ada aktivasi relu dan sigmoid </h4>

```python
modseq = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_dim=6),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(num, activation='sigmoid')
])
modseq.summary()
```

<h4> Proses Compile </h4>

```python
modseq.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.categorical_crossentropy,
    metrics=['accuracy']
)
```

<h4> disinilah proses training menggunakan cara fit </h4>

```python
trainingProcess = modseq.fit(
    x = xTrain,
    y = yTrain,
    validation_data = (xTest, yTest),
    epochs = 150,
    steps_per_epoch = 50,
    verbose = 2
)
```

<h4> dan terakhir simpan model </h4>

```python
modseq.save('model/seq_1.h5')
```

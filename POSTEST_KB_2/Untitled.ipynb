<h4> Mengimpor library pandas </h4>

```python
import pandas as pd
```

```python
df = pd.read_csv("CarRentalData.csv")
```

```python
df
```

<h3> ini adalah informasi, yaitu ada record, float, objek, ankga, dan null </h3>

```python
df.dtypes
```

<h4> Masukkan informasi di atas dalam dictionary dengan key sesuai nama informasinya </h4>

```python
JumlahRecord = len(df)
namaAttribute = df.columns
jumlahObjek = len(df.select_dtypes(include=['object']).columns)
jumlahAngka = len(df.select_dtypes(include=['int64', 'float64']).columns)
jumlahNULL = len(df.columns[df.isna().sum() > 0])

dictionary_no1 = {
    "Jumlah record"  : JumlahRecord,
    "Nama Attribute" : namaAttribute,
    "Jumlah Objek"   : jumlahObjek,
    "Jumlah Angka"   : jumlahAngka,
    "Jumlah NULL"    : jumlahNULL,
}

pd.Series(dictionary_no1)
```

<h4> Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka </h4>

<h4> Tampilkan Nilai rata-rata dengan fungsi mean </h4>

```python
df.select_dtypes(include=['int64' , 'float64']).mean()
```

<h4> Tampilkan Nilai Standar deviasi </h4>

```python
df.select_dtypes(include=['int64', 'float64']).std()    
```

<h4> Tampilkan Nilai Min</h4>

```python
df.select_dtypes(include=['int64', 'float64']).min()
```

<h4> Tampilkan Nilai Max </h4>

```python
df.select_dtypes(include=['int64', 'float64']).max()
```

<h4> Tampilkan Nilai Q1, Q2, dan Q3 dengan fungsi quantile </h4>

```python
df.select_dtypes(include=['int64', 'float64']).quantile([0.25,0.5,0.75])
```

<h3> jadi kita import dulu library pandas, matplotlib, seaborn </3>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("CarRentalData.csv")
```

<h3> Melihat dataset yang kita gunakan </h3>

```python
df
```

<h3> Menggunakan visualisasi bar <h/3>
<h3> lalu disini saya ada dua sumbu x dan y, yang dimana saya mengambil data atribut tipe kendaraan dan tahun kendaraan </h3>

```python
x = df['vehicle.year']
y = df['vehicle.type']
warna = '#696969'
```

```python
plt.bar(x, y, align="edge", width=0.6, color=warna)
plt.title("Kendaraan rental")
plt.xlabel("Banyak Tahun kendaraan")
plt.ylabel("Banyak Tipe Kendaraan")
plt.show()
```

<h3> Lalu untuk visualisasi yang kedua saya menggunakan grafik pie </h3>
<h3> disini bisa dilihat untuk banyaknya tiap jenis tipe kendaraan rental </h3>

```python
df['vehicle.type'].value_counts()
```

```python
labels = df['vehicle.type'].value_counts().rename_axis('vehicle.type').reset_index(name='Jumlah')
labels
```

```python
plt.figure(figsize=(7,7), dpi=200)
plt.pie(df['vehicle.type'].value_counts(), labels=labels['vehicle.type'], autopct='%.1f%%', startangle=65)
plt.show()

```
